**GET**:  ```<api-url>/aqueduct/analysis?geostore=<geostore>&wscheme=<weight scheme>```  
* accepted params: 
    * geostore: geostore id that represents a feature collection.
    * wscheme: weight scheme array as explained in [AQ_2_Weights](/notebooks/projects/wri-family/Aqueduct/lab/AQ_2_Weights.ipynb)
* response:  

```json
{
"data": [
    {"column1":"value",
    "column2":"value",
    ...
    },
    ...
    ]
}
``` 
* workflow:
    - Recive request
    - Validate params
    - Get geojson from geostore
    - Validate geojson as point collection (optional?)
    - Transform geojson feature collection into array of wkt
    - Format [query](/notebooks/projects/wri-family/Aqueduct/lab/AQ_2_Points_intersection.ipynb)
    - Send query to carto
    - Recive carto answer
    - Serialize and return microservice response
    

````python
%matplotlib inline
import geojson
import geopandas as gpd
from shapely.geometry import Point
````

````python
geo='''
{
  "type": "FeatureCollection",
  "features": [
    {
      "type": "Feature",
      "properties": {
      "test":1
      },
      "geometry": {
        "type": "Point",
        "coordinates": [
          72.421875,
          56.559482483762245
        ]
      }
    },
    {
      "type": "Feature",
      "properties": {
      "test":1
      },
      "geometry": {
        "type": "Point",
        "coordinates": [
          18.6328125,
          10.833305983642491
        ]
      }
    },
    {
      "type": "Feature",
      "properties": {
      "test":1
      },
      "geometry": {
        "type": "Point",
        "coordinates": [
          -109.6875,
          42.032974332441405
        ]
      }
    }
  ]
}

'''
````

````python
f = geojson.loads(geo)
wfs_gdf = gpd.GeoDataFrame.from_features(f.__geo_interface__['features'])
world = gpd.read_file(gpd.datasets.get_path('naturalearth_lowres'))
wfs_gdf.plot(ax=world.plot(alpha=1, color='grey'), figsize=(10, 6),
             marker='o', color='red', markersize=4);
````

````python
t = ['{0}'.format(Point(geometry["geometry"]["coordinates"]).wkt) for geometry in f.__geo_interface__['features']]    

query = 'SELECT ST_GeomFromText(unnest(Array{0}),4326) as the_geom'.format(str(t))
print(query)
    
````
