```python
import requests
import json
import pandas as pd
```

```python
url='https://api.openaq.org/v1/latest'
payload = {
    'limit':10000
}
r = requests.get(url, params=payload)
r.status_code
```

```python
pd.read_json(json.dumps(r.json()['results']),orient='records')
```

```python
from pandas.io.json import json_normalize
json_normalize(r.json()['results'], 'measurements', ['city','country','location', 'coordinates'])

```
